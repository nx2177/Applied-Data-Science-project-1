```python
import pandas as pd
file_path = "C:/Users/17376/Desktop/CYX/学习/5243 Applied DS/Project1/cleaned_data_v3.csv"
df = pd.read_csv(file_path)
```

```python

df["Margins_Percentage"] = df["Margins.of.Supply.Chain.Emission.Factors"] / df["Supply.Chain.Emission.Factors.with.Margins"].replace(0, float("nan"))
df["Margins_Percentage"].fillna(0, inplace=True)


def classify_industry(industry_title):
    industry_title = str(industry_title).lower()  
    if any(x in industry_title for x in ["cement", "steel", "oil", "coal"]):
        return "Energy-Intensive"
    elif any(x in industry_title for x in ["apparel", "luxury", "fashion", "software"]):
        return "High-Value"
    elif any(x in industry_title for x in ["consulting", "it", "finance", "law"]):
        return "Low-Emission"
    else:
        return "Other"

df["Industry_Category"] = df["2017.NAICS.Title"].apply(classify_industry)


df = pd.get_dummies(df, columns=["Industry_Category"], drop_first=False)


df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.histplot(df["Supply.Chain.Emission.Factors.with.Margins"], bins=30, kde=True)
plt.xlabel("Supply Chain Emission Factors (kg CO₂e/USD)")
plt.ylabel("Frequency")
plt.title("Distribution of Supply Chain Emission Factors")
plt.show()
```

```python
correlation_matrix = df[[
    "Supply.Chain.Emission.Factors.with.Margins",
    "Margins.of.Supply.Chain.Emission.Factors",
    "Margins_Percentage"
]].corr()

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Emission Factors")
plt.show()
```

```python
sns.scatterplot(
    x=df["Supply.Chain.Emission.Factors.without.Margins"],
    y=df["Margins.of.Supply.Chain.Emission.Factors"],
    alpha=0.5
)
plt.xlabel("Emissions Without Margins")
plt.ylabel("Margins")
plt.title("Comparison of Emissions Without and With Margins")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 选择聚类变量
X = df[["Supply.Chain.Emission.Factors.with.Margins", "Margins_Percentage"]]

# 数据标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 进行 K-Means 聚类
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X_scaled)

# 可视化聚类结果
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=df["Supply.Chain.Emission.Factors.with.Margins"],
    y=df["Margins_Percentage"],
    hue=df["Cluster"],
    palette="viridis"
)
plt.xlabel("Supply Chain Emission Factors")
plt.ylabel("Margins Percentage")
plt.title("Cluster Analysis of Industries Based on Emission Factors")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 画出三维散点图
scatter = ax.scatter(
    df["Supply.Chain.Emission.Factors.with.Margins"],
    df["Margins_Percentage"],
    df["Cluster"],
    c=df["Cluster"], cmap="viridis", alpha=0.7
)

ax.set_xlabel("Supply Chain Emission Factors")
ax.set_ylabel("Margins Percentage")
ax.set_zlabel("Cluster Group")
ax.set_title("3D Cluster Analysis of Supply Chain Emission Factors")

plt.colorbar(scatter)
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

plt.figure(figsize=(10, 6))
parallel_coordinates(df[["Cluster", "Supply.Chain.Emission.Factors.with.Margins", "Margins_Percentage"]], class_column="Cluster", colormap="viridis")
plt.title("Parallel Coordinates Plot of Cluster Characteristics")
plt.xlabel("Features")
plt.ylabel("Values")
plt.show()
```
